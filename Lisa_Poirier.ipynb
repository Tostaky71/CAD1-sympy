```python
#Q.1.1.

z, i, x, = symbols('z,i,x')

def f1(z,i):
    return z**2 + 6*z + (1-2*i)
A1 = solve(f1(z,i), z)
print("f1 =", A1)

def f2(z):
    return z**2 + 24*z + 37
A2 = solve(f2(z), z)
print("f2 =", A2)

def f3(x):
    return 2*sin((5*x)/2)-1
A3_1 = solve(f3(x), x)
A3_2 = solve(f3(x), sin((5*x)/2))
print("f3_x =", A3_1)
print("f3_sin =", A3_2)

def f4(x):
    return 3*x**2 + 1 - (x**2 +1)**3
A4 = solve(f4(x), x)
print("f4 =", A4)

def f5(x):
    return 4*cot(x)*(cos(x))**2 - cot(x)
A5_1 = solve(f5(x), x)
A5_2 = solve(f5(x), cos(x))
print("f5_x =", A5_1)
print("f5_cos =", A5_2)
```

```python
#Q.1.2.

def f6(x):
    return 2*sin(2*x)
print ("(f6)' =", f6(x).diff())
print ("(f6)'' =", (f6(x).diff()).diff())

def f7(x):
    return cos(x)/(cos(x)+1)
print ("(f7)' =", f7(x).diff())
print ("(f7)'' =", (f7(x).diff()).diff())

def f8(x):
    return (x**2 -1)/(x*((x-1)**2))
print ("(f8)' =", f8(x).diff())
print ("(f8)'' =", (f8(x).diff()).diff())
```

```python
#Q.1.3.

deriv_f6 = f6(x).diff()
integrate_f6 = integrate(deriv_f6 , x)
print ("For f6 :", integrate_f6)

deriv_f7 = f7(x).diff()
integrate_f7 = integrate(deriv_f7 , x)
print ("For f7 :",integrate_f7.simplify())

deriv_f8 = f8(x).diff()
integrate_f8 = integrate(deriv_f8 , x)
print ("For f8 :",integrate_f8)
```

```python
#Q.2.1.

Q,A,g,y = symbols('Q,A,g,y')

def E(Q,A,g,y):
    return (((Q/A)**2)/(2*g)) + y

E(Q,A,g,y)
```
