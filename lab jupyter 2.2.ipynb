```python
from numpy import *
from sympy import *
```

```python
t = symbols('t')
c1 = (cos(t), sin(t), (t,0, 2*pi))
c2 = (0.35+0.25*cos(t), 0.35+0.25*sin(t), (t,0, 2*pi))
c3 = (-0.35+0.25*cos(t), 0.35+0.25*sin(t), (t,0, 2*pi))
c4 = (cos(t)*2/3, sin(t)*2/3, (t, pi, 2*pi))

p = plotting.plot_parametric(c1, c2, c3, c4, show=False)
p.axis=False

p[0].line_color = '#501080'
p[1].line_color = '#ff0000'
p[2].line_color = '#ff0000'
p[3].line_color = '#008020'

p.show()
```

```python
t = symbols('t')
c1 = (cos(t), sin(t), (t,0, 2*pi))
c2 = (0.35+0.25*cos(t), 0.35+0.25*sin(t), (t,0, 2*pi))
c3 = (2-cos(t)*2/3, 2-sin(t)*2/3, (t,0, 2*pi))
c4 = (-cos(t)*2/3, -sin(t)*2/3, (t, pi, 2*pi))

p = plotting.plot_parametric(c1, c2, c3, c4, show=False)
p.axis=False

p[0].line_color = '#501080'
p[1].line_color = '#12510'
p[2].line_color = '#ff0000'
p[3].line_color = '#008020'

p.show()
```
